```python
'''
    Index numbers:
        Total fraction: 4
        
        Fraction Zero_1 : 5
        Fraction One_1 : 6
        
        Fraction Zero_2: 7
        Fraction One_2: 8
'''
def get_two_attr_diff(fontsize, labelsize, imagesize, dpi, line_width, ylim=None):
    
    greedy_diff_1 = np.abs(greedy[:, 5] - greedy[:, 6])
    greedy_diff_2 = np.abs(greedy[:, 7] - greedy[:, 8])
    
    fair_diff_1 = np.abs(fair_reclus[:, 5] - fair_reclus[:, 6])
    fair_diff_2 = np.abs(fair_reclus[:, 7] - fair_reclus[:, 8])
    
    Tsang_diff_1 = np.abs(np.round(Tsang_allfair[:, 5]/97, 2) - np.round(Tsang_allfair[:, 6]/344, 2))
    Tsang_diff_2 = np.abs(np.round(Tsang_allfair[:, 7]/282, 2) - np.round(Tsang_allfair[:, 8]/159, 2))
    
    fig, ax = plt.subplots(figsize=imagesize, dpi=dpi)
    

    ax.plot(n_seeds, greedy_diff_1, linestyle='solid', color='red', lw=line_width)
    ax.plot(n_seeds, greedy_diff_2, linestyle='dashed', color='red', lw=line_width)
    
    ax.plot(n_seeds, fair_diff_1, linestyle='solid', color='blue', lw=line_width)
    ax.plot(n_seeds, fair_diff_2, linestyle='dashed', color='blue', lw=line_width)
    
    ax.plot(n_seeds, Tsang_diff_1, linestyle='solid', color='green', lw=line_width)
    ax.plot(n_seeds, Tsang_diff_2, linestyle='dashed', color='green', lw=line_width)
    
    custom_lines = [Line2D([0], [0], color='red', lw=3),
                Line2D([0], [0], color='blue', lw=3),
                Line2D([0], [0], color='green', lw=3)]
    legend1 = plt.legend(custom_lines, ['Greedy', 'Fair-Emb', 'Tsang et al.'], bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
           ncol=2, mode="expand", borderaxespad=0.)
    plt.gca().add_artist(legend1)
    
    custom_lines2 = [Line2D([0], [0], linestyle='solid', lw=3), Line2D([0], [0], linestyle='dashed', lw=3)]
    legend2 = plt.legend(custom_lines2, ['Attr-1', 'Attr-2'], loc='upper left')
    plt.gca().add_artist(legend2)
    
    if ylim:
        ax.set_ylim([0, ylim])
    
    ax.set_axisbelow(True)
    ax.yaxis.grid(color='gray', linestyle='dashed')
    
    plt.xlabel('Number of Seeds', fontsize = labelsize, fontweight = 'bold')
    plt.ylabel('Diff in Fraction Influenced', fontsize = labelsize, fontweight = 'bold')
    plt.rcParams.update({'font.size': fontsize})
    plt.rcParams.update({'font.weight': 'bold'})
    plt.rcParams.update({'legend.fontsize' : '23'})
    plt.xticks(fontsize=labelsize)
    plt.yticks(fontsize=labelsize)
    
    fig.savefig('two-attr-diff' + '.pdf')        
```
